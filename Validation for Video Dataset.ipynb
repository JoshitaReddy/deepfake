```python
import os
import cv2
from ultralytics import YOLO
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix
```

```python
model = YOLO('C:/Users/joshi/OneDrive/Desktop/DeepFake/predict/bestAdam.pt')
```

```python
def predict(folder_path):
    img_results = []
    for filename in os.listdir(folder_path):
        if filename.lower().endswith(('.jpg')):
            results = model(folder_path + '/' + filename) 
            if results[0].probs.data[0] < results[0].probs.data[1] and results[0].probs.data[1] > 0.90:
                img_results.append('Real')
            elif results[0].probs.data[0] > results[0].probs.data[1] and results[0].probs.data[0] > 0.90:
                img_results.append('Fake')
    real_count = img_results.count('Real')
    fake_count = img_results.count('Fake')
    if real_count > fake_count:
        final_result = 'Real'
    else:
        final_result = 'Fake'
    return final_result
```

```python
def iterate_folders(root_folder):
    total_real_count = 0
    total_fake_count = 0
    for folder_name in os.listdir(root_folder):
        folder_path = os.path.join(root_folder, folder_name)
        if os.path.isdir(folder_path):
            if len(os.listdir(folder_path)) == 0:
                os.rmdir(folder_path)
                continue
            final_result = predict(folder_path)
            if final_result == 'Real':
                total_real_count += 1
            else:
                total_fake_count += 1
    return total_real_count, total_fake_count
```

```python
def normalize_confusion_matrix(confusion_matrix):
    row_sums = confusion_matrix.sum(axis=1, keepdims=True)
    normalized_matrix = confusion_matrix / row_sums
    return normalized_matrix

confusion_matrix_data = np.array([[121, 28], [18, 25]])
classes = ['Fake', 'Real']

normalized_matrix_data = normalize_confusion_matrix(confusion_matrix_data)

true_positives = confusion_matrix_data[0, 0]
false_positives = confusion_matrix_data[1, 0] 
false_negatives = confusion_matrix_data[0, 1]
true_negatives = confusion_matrix_data[1, 1]
total_samples = np.sum(confusion_matrix_data)

accuracy = (np.sum(true_positives) + true_negatives) / total_samples
precision = true_positives / (true_positives + false_positives)
recall = true_positives / (true_positives + false_negatives)
f1 = 2 * (precision * recall) / (precision + recall)

print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
print("Accuracy:", accuracy)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the confusion matrix values
conf_matrix = np.array([[146, 3], [4, 39]])

# Normalize the confusion matrix
conf_matrix_normalized = conf_matrix / conf_matrix.sum(axis=1)[:, np.newaxis]

# Define class labels
class_labels = ['Fake', 'Real']

# Plot the normalized confusion matrix
plt.figure(figsize=(6, 6))
plt.imshow(conf_matrix_normalized, interpolation='nearest', cmap=plt.cm.Blues)

cbar = plt.colorbar(shrink=0.72) 
tick_marks = np.arange(len(class_labels))
plt.xticks(tick_marks, class_labels)
plt.yticks(tick_marks, class_labels)

# Add numerical values inside the cells
thresh = conf_matrix_normalized.max() / 2.
for i, j in [(0, 0), (0, 1), (1, 0), (1, 1)]:
    plt.text(j, i, format(conf_matrix_normalized[i, j], '.2f'),
             ha="center", va="center",
             color="white" if conf_matrix_normalized[i, j] > thresh else "black")

plt.ylabel('Predicted Label')
plt.xlabel('True Label')
plt.tight_layout()
plt.show()
```
